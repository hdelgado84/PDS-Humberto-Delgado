Para ello analice señales determinísticas, por ejemplo una senoidal de **energía normalizada** y frecuencia $f_0$, y otras que varíen de forma más aleatoria y por lo tanto, representativa de una señal real, por ejemplo añadiendo ruido. Es decir que la señal *real* simulada será

donde la señal $n$ debe tener una energía de $1/10$.
           
Puede simular una señal continua (aproximadamente para los fines prácticos), mediante una señal muestreada a una $f^1_S$ muy alta en comparación con las $f_S$  que quiere estudiar, y un tipo de dato de doble precisión (double o float de 64 bits). De esta manera se asegura una grilla de cuantización (en tiempo y amplitud) mucho más densa que la que se va a simular. Se pide cuantizar a palabras de 4, 8 y 16 bits, obteniendo

$$ s_Q = \mathop{Q}_i \{s_R\} \quad i \in (4, 8, 16) $$

Siendo $\mathop{Q}_i$ la función **no lineal** responsable de la cuantización a *i-* bits de la señal real $s_R$. Describa el ruido de cuantización 

$$ e = s_Q - s_R$$

para las 3 situaciones antedichas siguiendo las siguientes consignas.

```python
#######################################
# Tu simulación que genere resultados #
#######################################

def ADC(señal, config):
    
    if config['modo'] == 'simple':
        b = config['bits']             #bits del ADC
        cuentas=2**b                   #canitdad de cuentas
        q=config['vref+']/cuentas      
        muestra=señal*(1/q)            #paso la señal a cuentas                     
        Cuantificado = np.around(muestra)*q  #cuantificaciòn,redondeo al
                                             #entero de la cuenta mas cercana 
        return Cuantificado
        
    elif config['modo'] == 'doble':
        b = config['bits']
        cuentas = 2**b    
        q= (config['vref+']-config['vref-'])/cuentas
        muestra=señal*(1/q)
        Cuantificado = np.around(muestra)*q
        
        return Cuantificado,q


Fs=1000
N=1000                  #cantidad de muestras de la señal
No=1000                 #cantidad de puntos de la DFT
F0=100
Ts=1/Fs
vnor=1/(np.sqrt(500))   #vnor para energía normalizada = 1

xs,ys=generador_senoidal(fs=Fs, f0=F0, N=1000, a0=vnor, p0=0)

sr =np.random.default_rng().uniform(-0.5,0.5,1000)

yr = np.sum(sr**2)
kn=np.sqrt(yr*10)    #calculo una K de normalización para energia de ruido 0.1
srn=(1/kn)*sr        #ruido normalizado a 0.1
yrn = np.sum(srn**2)

Sr= ys+srn           #Señal + ruido cuantificado


dftSr=DFT(Sr,No)

XSr=abs(dftSr)
Esrfs2=np.sum(XSr[0:No//2]**2)/No  #Energía de Señal + ruido hasta Fs2

Cfg_ADC={'bits' : 4,
         'vref+' : vnor,
         'vref-': -vnor,
         'bias':0,
         'modo':'doble',#modo "doble" para las vref+ y vref- "simple" Vref+
         
         }

Sc4,Q4 = ADC(Sr,Cfg_ADC)
#calculo del error
e4 = Sc4 - Sr
#energía de cuantización
dftsc4=DFT(Sc4, N=No)
XSc4=abs(dftsc4)
Esc4fs2=np.sum(XSc4[0:No//2]**2)/No
#energía de error
dfte4=DFT(e4, N=No)
Xe4=abs(dfte4)
Ee4fs2=np.sum(Xe4[0:No//2]**2)/No


Cfg_ADC={'bits' : 8,
         'vref+' : vnor,
         'vref-': -vnor,
         'bias':0,
         'modo':'doble',#modo "doble" para las vref+ y vref- "simple" Vref+
         
         }
Sc8,Q8 = ADC(Sr, Cfg_ADC)

e8= Sc8 - Sr
xe=np.arange(e8.size)
#energía de cuantización
dftsc8=DFT(Sc8, N=No)
XSc8=abs(dftsc8)
Esc8fs2=np.sum(XSc8[0:No//2]**2)/No
#energía de error
dfte8=DFT(e8, N=No)
Xe8=abs(dfte8)
Ee8fs2=np.sum(Xe8[0:No//2]**2)/No

Cfg_ADC={'bits' : 16,
         'vref+' : vnor,
         'vref-': -vnor,
         'bias':0,
         'modo':'doble',#modo "doble" para las vref+ y vref- "simple" Vref+
         
         }

Sc16,Q16 = ADC(Sr,Cfg_ADC)

e16 = Sc16 - Sr
#energía de cuantización
dftsc16=DFT(Sc16, N=No)
XSc16=abs(dftsc16)
Esc16fs2=np.sum(XSc16[0:No//2]**2)/No
#energía de error
dfte16=DFT(e16, N=No)
Xe16=abs(dfte16)
Ee16fs2=np.sum(Xe16[0:No//2]**2)/No


tus_resultados = [ ['$\sum_{f=0}^{f_S/2} \lvert S_R(f) \rvert ^2$', '$\sum_{f=0}^{f_S/2} \lvert S_Q(f) \rvert ^2$', '$\sum_{f=0}^{f_S/2} \lvert e(f) \rvert ^2$' ], 
                   ['',                                             '',                                             ''                              ], 
                   [Esrfs2, Esc4fs2, Ee4fs2], # <-- completar acá
                   [Esrfs2, Esc8fs2, Ee8fs2], # <-- completar acá
                   [Esrfs2, Esc16fs2, Ee16fs2], # <-- completar acá
                 ]
df = DataFrame(tus_resultados, columns=['Energía total', 'Energía total Q', 'Energía total $e$'],
               index=['$f_0$ \ expr. matemática', 
                      '', 
                      '4 bits', 
                      '8 bits', 
                      '16 bits'
                      ])
HTML(df.to_html())
```
